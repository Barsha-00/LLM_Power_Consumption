Q1. What was the average active power consumption in March 2007?

**Prompt: Given a pandas DataFrame named `df1` where the index is a datetime and the column `Global_active_power` contains energy consumption in kilowatts, write Python code to calculate the average active power consumption for the month of March 2007. give the result upto 2 decimal point**


```python
march_2007 = df1[(df1.index.month == 3) & (df1.index.year == 2007)]

# Calculate average active power consumption for March 2007
average_power = march_2007['Global_active_power'].mean()

print(f"Average Active Power Consumption for March 2007: {average_power:.2f} kW")
```

Q2. What hour of the day had the highest power usage on Christmas 2006?

**Prompt: Given a pandas DataFrame named `df1` where the index is a datetime and the column `Global_active_power`. write Python code to find the hour of the day that had the highest  power usage on Christmas 2006.**

```python
christmas_2006 = df1[(df1.index.month == 12) & (df1.index.day == 25) & (df1.index.year == 2006)]

# Group by hour of day and calculate mean power usage
hourly_power = christmas_2006.groupby(christmas_2006.index.hour)['Global_active_power'].mean()
#highest power usage
peak_hour = hourly_power.idxmax()
peak_power = hourly_power.max()

print(f"Hour with the highest power usage on Christmas 2006: {peak_hour}:00")
print(f"Highest power usage: {peak_power} kW")
```

Q3.Compare energy usage (Global_active_power) on weekdays vs weekends.

**Prompt: Given a pandas DataFrame named `df1` where the index is a datetime and the column `Global_active_power`.  write Python code to compare  energy usage on weekdays vs weekends.**


```python
import matplotlib.pyplot as plt
df1['Day_Type'] = df1.index.map(lambda x: 'Weekend' if x.weekday() >= 5 else 'Weekday')

# Group by day type and calculate mean energy usage
energy_usage = df1.groupby('Day_Type')['Global_active_power'].mean()

# Print the mean energy usage for weekdays and weekends
print(energy_usage)

# Plot the mean energy usage for weekdays and weekends
energy_usage.plot(kind='bar')
plt.xlabel('Day Type')
plt.ylabel('Mean Energy Usage (kW)')
plt.title('Comparison of Energy Usage on Weekdays vs Weekends')
plt.show()
```

Q4. Find days where energy consumption exceeded 5 kWh.

**Prompt: Given a pandas DataFrame named `df1` where the index is a datetime and the column `Global_active_power` contains energy consumption in kilowatts.  write Python code to find all days where the total energy consumption exceeded 5 kilowatt-hours.**

```python
daily_energy = df1['Global_active_power'].resample('D').sum()

# Find days where total energy consumption exceeded 5 kilowatt-hours
high_energy_days = daily_energy[daily_energy > 5]
print(high_energy_days)
```

Q5. Plot the energy usage trend for the first week of January 2007.

**Prompt: Given a pandas DataFrame named `df1` where the index is a datetime and the column `Global_active_power`.write Python code to plot the energy usage trend for the first week of January 2007.**

```python
import matplotlib.pyplot as plt

# Filter data for the first week of January 2007
first_week_jan_2007 = df1[(df1.index.year == 2007) & (df1.index.month == 1) & (df1.index.day <= 7)]

# Plot energy usage trend
plt.figure(figsize=(10, 6))
plt.plot(first_week_jan_2007.index, first_week_jan_2007['Global_active_power'])
plt.xlabel('Date')
plt.ylabel('Global Active Power (kW)')
plt.title('Energy Usage Trend - First Week of January 2007')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Q6. Find the average voltage for each day of the first week of February 2007.

**Prompt: Given a pandas DataFrame named `df1` where the index is a datetime and the  column `Voltage`, write Python code to calculate the average voltage for each day of the first week of February 2007.**



```python
first_week_feb_2007 = df1[(df1.index.year == 2007) & (df1.index.month == 2) & (df1.index.day <= 7)]

# Calculate average voltage for each day
average_voltage = first_week_feb_2007.groupby(first_week_feb_2007.index.date)['Voltage'].mean()

print(average_voltage)
```

Q7. What is the correlation between global active power and sub-metering values?

**Prompt:Given a pandas DataFrame `df1` with columns `Global_active_power`, `Sub_metering_1`, `Sub_metering_2`, and `Sub_metering_3`, write Python code to compute and display the correlation matrix between global active power and sub-metering values.**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Compute the correlation matrix
correlation_matrix = df1[['Global_active_power', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']].corr()

# Display the correlation matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix between Global Active Power and Sub-metering Values')
plt.show()

# Display the correlation matrix as a table
print(correlation_matrix)
```
